Write a Python program to check whether a number is prime or not.

```python
n=int(input('enter a positive number'))
if(n == 2):
      print(n,'is prime')
if n > 1:
  for i in range (2,n):
    if(n % i == 0):
      print(n,'is a not a prime')
      break
    else:
      print(n,'is prime')
else:
  print(n,'is not prime')
```
